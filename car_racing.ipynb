```python
#Display the correlation between numerical features

import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns
numerical_data = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr = numerical_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Show how the number of laps correlates with the number of pitstops
plt.figure(figsize=(10, 8))
sns.heatmap(df.pivot_table(index='Laps', columns='Pitstops', aggfunc='size', fill_value=0), cmap='YlGnBu', annot=True, fmt='d')
plt.title("Laps vs Pitstops Heatmap")
plt.show()
```

```python
#Visualize the average Best Lap Kph for each Category
avg_speed = df.groupby('Category')['Best Lap Kph'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.heatmap(avg_speed.pivot_table(index='Category', values='Best Lap Kph'), annot=True, cmap='viridis', fmt=".1f")
plt.title("Average Speed by Category")
plt.show()
```

```python
#Analyze the average number of laps completed by cars grouped by Category and Hour

laps_by_category_hour = df.pivot_table(index='Category', columns='Hour', values='Laps', aggfunc='mean', fill_value=0)

plt.figure(figsize=(12, 6))
sns.heatmap(laps_by_category_hour, annot=True, cmap='coolwarm', fmt=".1f")
plt.title("Average Laps Completed by Category and Hour")
plt.show()
```

```python
#Visualize the average number of pitstops for teams across different categories.

pitstops_team_category = df.pivot_table(index='Team', columns='Category', values='Pitstops', aggfunc='mean', fill_value=0)

plt.figure(figsize=(14, 10))
sns.heatmap(pitstops_team_category, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title("Average Pitstops by Team and Category")
plt.show()
```
